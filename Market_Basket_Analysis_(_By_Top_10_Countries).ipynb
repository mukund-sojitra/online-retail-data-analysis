```python
#Loading neccesary packages
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import *
```

```python
path = r"/content/gdrive/MyDrive/Sem 4 project/Online Retail_1.csv"
data = pd.read_csv(path)
data.head()
```

```python
#Top 10 Countries by Transaction Counts
data['Country'].value_counts().head(10)
```

# Exploring United Kingdom

```python
#Separating transactions for United kingdom
United_kingdom = (data[data['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
United_kingdom.head()
```

```python
#converting all positive vaues to 1 and everything else to 0
def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

United_kingdom_item_sets = United_kingdom.applymap(my_encode_units)
#United_kingdom_item_sets.drop('POSTAGE', inplace=True, axis=1) #Remove "postage" as an item
```

```python
United_kingdom_item_sets.T
```

```python
# !git clone https://github.com/jroakes/querycat.git
# !pip install -r querycat/requirements_colab.txt

import querycat
```

```python
# from mlxtend.frequent_patterns import fpgrowth

# United_kingdom_frequent_itemsets = fpgrowth(United_kingdom_item_sets, min_support=0.01, use_colnames=True,verbose=0)
# United_kingdom_frequent_itemsets


United_kingdom_frequent_itemsets = querycat.Categorize(United_kingdom_item_sets, United_kingdom_item_sets.columns.to_list(), min_support=10,  alg='fpgrowth ')
United_kingdom_frequent_itemsets
```

```python
#help(fpgrowth)
#help(apriori)
```

```python
United_kingdom_rules = association_rules(United_kingdom_frequent_itemsets, metric="confidence", min_threshold=0.1)
United_kingdom_rules
```

```python
def Get_recommendations_UK(product):
   
    UK_Rules = United_kingdom_rules[United_kingdom_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
    
    return UK_Rules
    
    
```

```python
def Get_recommendations_by_country(product,country):
    
    if country == "United_Kingdom":
        UK_Rules = United_kingdom_rules[United_kingdom_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return UK_Rules
    
    elif country == "Germany":
        Germany_Rules = Germany_rules[Germany_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Germany_Rules
    
    elif country == "France":
        France_Rules = France_rules[France_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return France_Rules
    
    elif country == "EIRE":
        EIRE_Rules = EIRE_rules[EIRE_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return EIRE_Rules
    
    elif country == "Spain":
        Spain_Rules = Spain_Rules = Spain_rules[Spain_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Spain_Rules
    
    elif country == "Netherlands":
        Netherlands_Rules = Netherlands_rules[Netherlands_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Netherlands_Rules
    
    elif country == "Belgium":
        Belgium_Rules = Belgium_rules[Belgium_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Belgium_Rules
    
    elif country == "Switzerland":
        Switzerland_Rules = Switzerland_rules[Switzerland_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Switzerland_Rules
    
    elif country == "Portugal":
        Portugal_Rules = Portugal_rules[Portugal_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Portugal_Rules
    
    else:
        Australia_Rules = Australia_rules[Australia_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False)
        return Australia_Rules
    
    
```

```python
Get_recommendations_by_country("PARTY BUNTING","Australia")
```

```python
United_kingdom_rules['Country'] = "United Kingdom"
Germany_rules['Country'] = "Germany"
France_rules['Country'] = "France"
EIRE_rules['Country'] = "EIRIE"
Spain_rules['Country'] = "Spain"
Netherlands_rules['Country'] = "Netherlands"
Belgium_rules['Country'] = "Belgium"
Switzerland_rules['Country'] = "Switzerland"
Portugal_rules['Country'] = "Portugal"
Australia_rules['Country'] = "Australia"
```

```python
All_Rules = pd.concat([United_kingdom_rules, Germany_rules,France_rules,
                   EIRE_rules,Spain_rules,Netherlands_rules,Belgium_rules,
                  Switzerland_rules,Portugal_rules,Australia_rules])
All_Rules.head()
```

```python
def Get_recommendations(product):
   
    Rules = All_Rules[All_Rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False).head(5)
    
  
    return Rules
    
    
```

```python
Get_recommendations('6 RIBBONS RUSTIC CHARM')
```
